### Exercício 5
Explique o que acontece no comando `modelo.fit(X, y)`.

Treina o modelo utilizando as features (X) para aprender a prever o target (y).

### Desafio
Crie um modelo que responda: **Um aluno que estudou 5 horas será aprovado?**

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = {
    "study_time": [1, 2, 3, 4, 5, 6],
    "test_results": [0, 0, 0, 1, 1, 1]
}

df = pd.DataFrame(data)

X = df[["study_time"]]
y = df["test_results"]

model_tree = DecisionTreeClassifier()
model_tree.fit(X, y)

# Ajustado para prever exatamente as 5 horas solicitadas
new_data = pd.DataFrame({"study_time": [5]})
prediction = model_tree.predict(new_data)

if prediction[0]:
  print("Pass")
else:
  print("Not Pass")
```
